```python
# 스택을 이용한 후위 표기법 계산기 만들기
# stack

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class Stack:
    def __init__(self):
        self.head = None

    def is_empty(self):
        if not self.head:
            return True
        return False

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def pop(self):
        if self.is_empty():
            return None
        ret_data = self.head.data
        self.head = self.head.next
        return ret_data
            
    def peek(self):
        if self.is_empty():
            return None
        return self.head.data

if __name__ == '__main__':
    s = Stack()
    s.push(3)
    s.push(4)
    s.push(5)

    print('peek : {}'.format(s.peek()))

    while not s.is_empty():
        print(s.pop())


class Calculator:
    def __init__(self, org_exp):
        self.org_exp = org_exp.replace(' ', '')
        self.postfix_exp = None
    
    def get_weight(self, oprt):
        if oprt == '*' or oprt == '/':
            return 9
        elif oprt == '+' or oprt == '-':
            return 7
        elif oprt == '(':
            return 5
        else:
            return None 

    def convert_to_postfix(self):
        exp_list = []
        oprt_stack = Stack()

        for ch in self.org_exp:
            if ch.isdigit():
                exp_list.append(ch)
            else:
                if ch == '(' or oprt_stack.is_empty():
                    oprt_stack.push(ch)

                elif ch == ')':
                    op = oprt_stack.pop()
                    while op != '(':
                        exp_list.append(op)
                        op = oprt_stack.pop()
                elif self.get_weight(ch) > self.get_weight(oprt_stack.peek()):
                    oprt_stack.push(ch)
                else:
                    while self.get_weight(ch) <= self.get_weight(oprt_stack.peek()):
                        exp_list.append(oprt_stack.pop())
                    oprt_stack.push(ch)

        while oprt_stack:
            exp_list.append(oprt_stack.pop())

        self.postfix_exp = ''.join(exp_list)
        return self.postfix_exp

    def calc_two_oprd(self, oprd1, oprd2, oprt):
        if oprt == '+':
            return oprd1 + oprd2
        elif oprt == '-':
            return oprd1 - oprd2
        elif oprt == '*':
            return oprd1 * oprd2
        elif oprt == '/':
            return oprd1 // oprd2

    def calculate(self):
        oprt_stack = Stack()

        for ch in self.postfix_exp:
            if ch.isdigit():
                oprt_stack.push(int(ch))
            else:
                oprd2 = oprd_stack.pop()
                oprd1 = oprd_stack.pop()
                oprd_stack.push(self.calc_two_oprd(oprd1, oprd2, ch))
        return oprd_stack.pop()

if __name__ == '__main__':
    a = input('수식을 입력하세요: ')  
    calc = Calculator(a)
    print(calc.org_exp)
    calc.convert_to_postfix()
#     calc.calculate()
#     calc = calc.convert_to_postfix()
#     print(calc.calculate())
```
