```python
import urllib
from bs4 import BeautifulSoup
```

```python
query = input("what do you want to find??")
url = "https://search.naver.com/search.naver?where=post&query=" + query 

# url = "https://search.naver.com/search.naver?where=post&query=cat"
html = urllib.request.urlopen(url)
source = html.read()
html.close()
```

```python
soup = BeautifulSoup(source, "html.parser")
ul = soup.find(id="elThumbnailResultArea")
print(ul)
```

```python
all_li = ul.find_all("a", attrs={"class": "sh_blog_title"})

# print(all_li)
for target in all_li:

    link = target.get('href')
    text = target.get('title')
    print(link, text)
```
